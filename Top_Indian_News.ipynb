## We will be scraping for article titles from India's top newspapers

Every time you open the Jupyter notebook, just click 'Run All' under the 'Cell' dropdown list and we will have scraped over 120 articles in one day and written it into 2 CSV files. As a python notebook, just run this file after commenting out the <i>print</i> statements in the numerous functions defined below so that your terminal will look less cluttered.

```python
#Importing all the necessary libraries for web-scraping
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from datetime import date, timedelta
```

<b>Most of the headlines cover COVID-19 related news extensively, but in case the newspaper didn't cover COVID-19 related news extensively, I've visited the newspapers' "Coronavirus" section and scraped the top articles from there.</b>
<br />
<li> The Times of India, India's largest selling newspaper</li>
<li> The Hindu</li>
<li> India Today</li>
<li> The News' Minute, a website that provides short, 1-minute-readable news with primary focus on the 5 southern Indian states</li>
<li> Indian Express</li>

```python
url1 = "https://timesofindia.indiatimes.com/home/headlines"
page_request1 = requests.get(url1)
data1 = page_request1.content
soup1 = BeautifulSoup(data1,"html5lib")
list_titles=[]

url2 = "https://www.thehindu.com/"
page_request2 = requests.get(url2)
data2 = page_request2.content
soup2 = BeautifulSoup(data2,"html5lib")
list_titles2=[]

url3 = "https://www.indiatoday.in/"
page_request3 = requests.get(url3)
data3 = page_request3.content
soup3 = BeautifulSoup(data3,"html5lib")
list_titles3=[]

url4 = "https://www.thenewsminute.com/section/coronavirus"
page_request4 = requests.get(url4)
data4 = page_request4.content
soup4 = BeautifulSoup(data4,"html5lib")
list_titles4=[]

url5 = "https://indianexpress.com/"
page_request5 = requests.get(url5)
data5 = page_request5.content
soup5 = BeautifulSoup(data5,"html5lib")
list_titles5=[]
```

<b> Let us define functions for each section of the newspaper. Sometimes the top headlines will be few in number, hence, the next set of top articles will have to be scraped for that day. Whenever I have felt the inadequacy of the number of titles, that newspaper has two functions defined for the number of top articles to exceed 10.</b>

```python
def timesofindia1():
    counter=0
    divitag=soup1.find('div', class_= 'top-newslist')
    for divtag in divitag:
        if (counter <= 6):
            for spantag in divtag.find_all('span', class_='w_tle'):
                counter = counter + 1
                print(spantag.find('a')['title'])
                list_titles.append(spantag.find('a')['title'])
                
def timesofindia2():
    counter=0
    divitag=soup1.find('div', class_= 'headlines-list')
    for divtag in divitag:
        if (counter <= 12):
            for spantag in divtag.find_all('span', class_='w_tle'):
                counter = counter + 1
                print(spantag.find('a')['title'])
                list_titles.append(spantag.find('a')['title'])

def thehindu():
    counter=0
    divitag=soup2.find('div', class_= 'row footrendsec')
    for divtag in divitag:
        for ahref in divitag.find_all('a'):
            if (counter < 10):
                counter = counter + 1
                print(ahref.get_text())
                list_titles2.append(ahref.get_text())

def indiatoday_topfeatured():
    divitag=soup3.find('div', class_= 'featured-post featured-post-first')
    print(divitag.h2.get_text())
    list_titles3.append(divitag.h2.get_text())

def indiatoday_topstory():
    divtag=soup3.find('ul', class_= 'itg-listing')
    for n in range(len(divtag)):
        if(divtag.find_all('li')[n].a.get_text()=='Covid-19 Tracker: State-wise data, daily trends, patients cured, deaths and more'):
            continue
        else:
            print(divtag.find_all('li')[n].a.get_text())
            list_titles3.append(divtag.find_all('li')[n].a.get_text())
        
def newsminute_topstory():
    divtag=soup4.find('h2', class_= 'article-title')
    print(divtag.a.text)
    list_titles4.append(divtag.a.get_text())

def newsminute_toparticles():
    for k in range (3):
        divtag=soup4.find_all('div', class_= 'row article-group-bottom')[k]
        for n in range(len(divtag.find_all('a'))):
            if (n%2!=0):
                print(divtag.find_all('a')[n].text)
                list_titles4.append(divtag.find_all('a')[n].text)
                
def indianexpress_toparticles():
    divtag=soup5.find('div', class_= 'ie-first-story')
    if((divtag.find('h1'))!=None):
        print(divtag.find('h1').text)
        list_titles5.append(divtag.find('h1').text)
    else:
        print(divtag.find('h2').text)              #sometimes, the top title is NOT present in the "h1" tag, but rather it is present in the "h2" tag
        list_titles5.append(divtag.find('h2').text)
    
    if(soup5.find('div', class_= 'news-breaking event-track-class')!=None):
        divtag=soup5.find('div', class_= 'news-breaking event-track-class')
        for n in range(len(divtag.find_all('li'))):
            print(divtag.find_all('li')[n].text)
            list_titles5.append(divtag.find_all('li')[n].text)
            
    if(soup5.find('div', class_= 'top-news')!=None):  
        divtag=soup5.find('div', class_= 'top-news')
        for n in range(len(divtag.find_all('h3'))):
            print(divtag.find_all('h3')[n].text)
            list_titles5.append(divtag.find_all('h3')[n].text)
        
        
def indianexpress_corona():     
    divtag=soup5.find('div', class_='left-panel')
    print(divtag.h3.text)
    
    for k in range(2):
        print(soup5.find_all('h5')[k].text)
        
    divtag=soup5.find('div', class_= 'right-panel')
    for n in range(len(divtag.find_all('h4'))):
        print((divtag.find_all('h4')[n].text))      
        list_titles5.append(divtag.find_all('h4')[n].text) 
    
    
```

```python
#Executing all the above defined functions
if __name__ == "__main__":
    timesofindia1()
    timesofindia2()
    thehindu()
    indiatoday_topfeatured()
    indiatoday_topstory()
    newsminute_topstory()
    newsminute_toparticles()
    indianexpress_toparticles()
    indianexpress_corona()
```

### Creating a dataframe to store the names of the newspapers, the date of scraping along with the top article titles that we have just obtained

```python
newspaper1=['Times of India' for i in range(len(list_titles))]
newspaper2=['The Hindu' for i in range(len(list_titles2))]
newspaper3=['India Today' for i in range(len(list_titles3))]
newspaper4=['The News Minute' for i in range(len(list_titles4))]
newspaper5=['The Indian Express' for i in range(len(list_titles5))]
newspaper=newspaper1+newspaper2+newspaper3+newspaper4+newspaper5
list=list_titles+list_titles2+list_titles3+list_titles4+list_titles5
dates=[(date.today()-timedelta(days=0)) for d in range(len(list_titles)+len(list_titles2)+len(list_titles3)+len(list_titles4)+len(list_titles5))]
#print(dates)
df_show_info = pd.DataFrame()
df_show_info['Newspaper']=newspaper
df_show_info['Date']=dates
df_show_info['Article Title']=list
```

```python
df_show_info
```

### Let us now write all of the data here into a CSV file

```python
df_show_info.to_csv('news.csv', mode='a', index=False, header=False)
```

### The following section from now includes only top news titles from the 3 most famous English newspapers in Andhra Pradesh. The most famous newspapers are infact, the local ones in Telugu, but since we cannot mine those webpages, we'll have to stick to English newspapers

Essentially 2 physical newspapers that are circulated and 'The News Minute' from before have been scraped for titles-
<li>Deccan Chronicle</li>
<li>New Indian Express, Indian Express's southern counterpart</li>
<li>The News Minute</li>

```python
#Andhra News
from fake_useragent import UserAgent

urla1 = "https://www.deccanchronicle.com/"
page_requesta1 = requests.get(urla1)
dataa1 = page_requesta1.content
soupa1 = BeautifulSoup(dataa1,"html5lib")
list_titlesa1=[]

ua = UserAgent()
header = {'User-Agent':str(ua.chrome)}
urla2 = "https://www.newindianexpress.com/states/andhra-pradesh"
page_requesta2 = requests.get(urla2, headers=header)
dataa2 = page_requesta2.content
soupa2 = BeautifulSoup(dataa2,"html5lib")
list_titlesa2=[]

urla3 = "https://www.thenewsminute.com/section/Andhra%20Pradesh"
page_requesta3 = requests.get(urla3)
dataa3 = page_requesta3.content
soupa3 = BeautifulSoup(dataa3,"html5lib")
list_titlesa3=[]
```

```python
def deccanchronicle_topstory():
    divtag=soupa1.find_all('div', class_= 'col-sm-12 noPadding stry-top-big-a')
    for n in range(len(divtag)):
        print(divtag[n].a.h2.text)
        list_titlesa1.append(divtag[n].a.h2.text)

def deccanchronicle_toparticles():
    divtag=soupa1.find('div', class_= 'col-sm-5 tsSmall')
    for n in range(len(divtag.find_all('h3'))):
        print(divtag.find_all('h3')[n].text)
        list_titlesa1.append(divtag.find_all('h3')[n].text)
        
        
def newindianexpress_topstory():
    divtag=soupa2.find_all('h4', class_= 'subtopic')
    for n in range(len(divtag)):
        print(divtag[n].text)
        list_titlesa2.append(divtag[n].text)

def newindianexpress_toparticles():
    divtag=soupa2.find_all('div', class_= 'search-row_type')
    for n in range(len(divtag)):
        print(divtag[n].find('h4').text)
        list_titlesa2.append(divtag[n].find('h4').text)
        
def newsminutea_topstory():
    divtag=soupa3.find('h2', class_= 'article-title')
    print(divtag.a.text)
    list_titlesa3.append(divtag.a.text)

def newsminutea_toparticles():
    for k in range (3):
        divtag=soupa3.find_all('div', class_= 'row article-group-bottom')[k]
        for n in range(len(divtag.find_all('a'))):
            if (n%2!=0):
                print(divtag.find_all('a')[n].text)
                list_titlesa3.append(divtag.find_all('a')[n].text)
```

```python
#Executing all the above defined functions
if __name__ == "__main__":
    deccanchronicle_topstory()
    deccanchronicle_toparticles()
    newindianexpress_topstory()
    newindianexpress_toparticles()
    newsminutea_topstory()
    newsminutea_toparticles()
```

### Creating a dataframe to store the names of the newspapers, the date of scraping along with the top article titles that we have just obtained

```python
newspapera1=['Deccan Chronicle' for i in range(len(list_titlesa1))]
newspapera2=['New Indian Express' for i in range(len(list_titlesa2))]
newspapera3=['The News Minute' for i in range(len(list_titlesa3))]
newspapera=newspapera1+newspapera2+newspapera3
lista=list_titlesa1+list_titlesa2+list_titlesa3
datesa=[(date.today()-timedelta(days=0)) for d in range(len(list_titlesa1)+len(list_titlesa2)+len(list_titlesa3))]
#print(lista)
dfa_show_info = pd.DataFrame()
dfa_show_info['Newspaper']=newspapera
dfa_show_info['Date']=datesa
dfa_show_info['Article Title']=lista
```
